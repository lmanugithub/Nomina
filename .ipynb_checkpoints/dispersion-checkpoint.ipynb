```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
pd.options.display.float_format = '$ {:,.2f}'.format
```

### Function to read files txt

```python
def lectura(ruta, pattern):

    # list comprehension to read all files in a path
    files = [file for file in os.listdir(ruta)]

    # creating empty list
    importe = []
    nombre = []
    archivo = []

    # the first loop is created to read files in a path
    for file in files:
        f = open(ruta+file, "r", encoding="utf8")

        # the second one is created to read line in the file
        for line in f:
          resultado = re.match(pattern, line)
          if resultado:
            importe.append(resultado.group(1))
            nombre.append(resultado.group(2))
            archivo.append(file)
        f.close()
    return (nombre, importe, archivo)
```

```python
def lectura_s(ruta, pattern):

    # list comprehension to read all files in a path
    files = [file for file in os.listdir(ruta)]

    # creating empty list
    importe = []
    nombre = []
    archivo = []
   
    # the first loop is created to read files in a path
    for file in files:
        f = open(ruta+file, "r", encoding="utf8")

        # the second one is created to read line in the file
        for line in f:
          resultado = re.match(pattern, line)
          if resultado:
            name_1 = resultado.group(1)
            name_2 = resultado.group(2)
            name_3 = resultado.group(3)
            all_name = name_3.strip()+' '+name_1.strip()+' '+name_2.strip()
            importe.append(resultado.group(4))
            nombre.append(all_name)
            archivo.append(file)
        f.close()
    return (nombre, importe, archivo)
```

### First, I want to read all files of PEMs

```python
# It is needed a path and a pattern (RegEx)
ruta = 'otros/archivos/pem_bancomer/'
pattern = re.compile(r'^[\d]{9}[\s]{16}[\d]{12}[\s]{10}([\d]{15})([A-Za-z\s]*).*$')
```

```python
listas = lectura(ruta=ruta, pattern=pattern)

df_pem = pd.DataFrame({'Nombre':listas[0],
                  'Importe':listas[1],
                  'Archivo':listas[2]})
df_pem['Importe'] = df_pem['Importe'].astype('int')
df_pem['Importe'] = df_pem['Importe'].apply(lambda x: x/100)
df_pem.head()
```

```python
# It is needed a path and a pattern (RegEx)
ruta = 'otros/archivos/pem_santander/'
pattern = re.compile(r'^[\d]{7,8}[\s]{5,6}([A-Z\s]{30})([A-Z\s]{20})([A-Z\s]{30})[\d]{11}[\s]{5}([\d]{18})05$')
```

```python
listas = lectura_s(ruta=ruta, pattern=pattern)

df_pem_s = pd.DataFrame({'Nombre':listas[0],
                  'Importe':listas[1],
                  'Archivo':listas[2]})
df_pem_s['Importe'] = df_pem_s['Importe'].astype('int')
df_pem_s['Importe'] = df_pem_s['Importe'].apply(lambda x: x/100)
df_pem_s.head()
```

```python
df_pem = pd.concat([df_pem,df_pem_s])
df_pem
```

```python
df_pem.groupby('Archivo').sum().reset_index()
```

```python
df_pem['Nombre'] = df_pem['Nombre'].str.strip()
```

```python
pem = pd.read_excel('./otros/archivos/pem.xlsx')
pem.head()
```

```python
df_pem = df_pem.merge(pem, how='outer', left_on='Nombre', right_on='Nombre')
```
