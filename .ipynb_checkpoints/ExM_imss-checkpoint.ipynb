```python
import pandas as pd
from cuotasimss import CuotasImss
from imss import CalculoIMSSObrero, CalculoIMSSPatron
from isr import ISR
from datetime import datetime
from sbc import SBC
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_excel('./data/plantilla_laboral2.xlsx', sheet_name='Hoja1') #2022-10-27 incremento
df.head()
```

```python
df['Fecha Ingreso'] = pd.to_datetime(df['Fecha Ingreso'], format='%Y-%m-%d')
```

```python
date_2022 = pd.to_datetime('2022/12/31')
defoult = 30.4
dias = 31
dias_b = 61
```

## Funciones Mensuales

```python
def cuota_fija( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.cuotas_pres_especie_enf_mat_cuot()/defoult*dias
```

```python
def excedente_patron( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.cuotas_pres_especie_enf_mat()/defoult*dias
```

```python
def excedente_obrero(x):
    calculo = CalculoIMSSObrero()
    calculo.set_sdi(x)
    return calculo.cuotas_pres_especie_enf_mat()/defoult*dias
```

```python
# prestación en dinero
def pd_patron( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.cuotas_pres_dinero_enf_mat()/defoult*dias
```

```python
# prestación en dinero
def pd_obrero(x):
    calculo = CalculoIMSSObrero()
    calculo.set_sdi(x)
    return calculo.cuotas_pres_dinero_enf_mat()/defoult*dias
```

```python
# gastos medicos y pensiones
def gmp_patron( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.gastos_medicos_pencionados()/defoult*dias
```

```python
# gastos medicos y pensiones
def gmp_obrero(x):
    calculo = CalculoIMSSObrero()
    calculo.set_sdi(x)
    return calculo.gastos_medicos_pencionados()/defoult*dias
```

```python
# riesgo de trabajo
def riesgo_trabajo( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.riesgo_de_trabajo()/defoult*dias
```

```python
# invalidez y vida
def iv_patron( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.cuota_invalidez()/defoult*dias
```

```python
# invalidez y vida
def iv_obrero(x):
    calculo = CalculoIMSSObrero()
    calculo.set_sdi(x)
    return calculo.cuota_invalidez()/defoult*dias
```

```python
# guarderia y prestaciones sociales
def guarderia_ps( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.guarderia()/defoult*dias
```

## Funciones bimestrales

```python
# retiro
def retiro( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.retiro()/defoult*dias_b
```

```python
# Cesantía en Edad Avanzada y Vejez
def rcv_patron( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.cuota_cesantia_vejez()/defoult*dias_b
```

```python
# Cesantía en Edad Avanzada y Vejez
def rcv_obrero(x):
    calculo = CalculoIMSSObrero()
    calculo.set_sdi(x)
    return calculo.cuota_cesantia_vejez()/defoult*dias_b
```

```python
# infonavit
def infonavit_patron( x, date):
    calculo = CalculoIMSSPatron()
    calculo.set_fecha_base(date)
    calculo.set_sdi(x)
    return calculo.infonavit()/defoult*dias_b
```

```python
df['SBC'] = df.apply(lambda x: SBC.funcion_sbc(sueldo = x['Sueldo mensual'], vales = x['Vales'], a = x['Fecha Ingreso'], b = date_2022), axis=1)
df['cuota fija'] = df.apply(lambda x: cuota_fija( x= x['SBC'], date =  date_2022), axis=1 )
df['excedente patron'] = df.apply(lambda x: excedente_patron( x= x['SBC'], date =  date_2022), axis=1 )
df['excedente obrero'] = df.apply(lambda x: excedente_obrero( x= x['SBC']), axis=1 )
df['PD patron'] = df.apply(lambda x: pd_patron( x= x['SBC'], date =  date_2022), axis=1 )
df['PD obrero'] = df.apply(lambda x: pd_obrero( x= x['SBC']), axis=1 )
df['GMP patron'] = df.apply(lambda x: gmp_patron( x= x['SBC'], date =  date_2022), axis=1 )
df['GMP obrero'] = df.apply(lambda x: gmp_obrero( x= x['SBC']), axis=1 )
df['riesgo de trabajo'] = df.apply(lambda x: riesgo_trabajo( x= x['SBC'], date =  date_2022), axis=1 )
df['IV patron'] = df.apply(lambda x: iv_patron( x= x['SBC'], date =  date_2022), axis=1 )
df['IV obrero'] = df.apply(lambda x: iv_obrero( x= x['SBC']), axis=1 )
df['Guarderia y PS'] = df.apply(lambda x: guarderia_ps( x= x['SBC'], date =  date_2022), axis=1 )
```
