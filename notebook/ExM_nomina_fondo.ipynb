```python
import pandas as pd
from cuotasimss import CuotasImss
from imss import CalculoIMSSObrero, CalculoIMSSPatron
from isr import ISR
from datetime import datetime
from sbc import SBC
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_excel('./data/plantilla_laboral2.xlsx', sheet_name='Hoja1') #2022-10-27 incremento
df.head()
```

```python
df['Fecha Ingreso'] = pd.to_datetime(df['Fecha Ingreso'], format='%Y-%m-%d')
```

```python
df_fondo = df.copy(deep=True)
```

```python
date_2022 = pd.to_datetime('2023/12/31')
```

```python
df['SD'] = df['Sueldo mensual'].apply(lambda x: x / 30)
df['SBC'] = df.apply(lambda x: SBC.funcion_sbc(sueldo = x['Sueldo mensual'], vales = x['Vales'], a = x['Fecha Ingreso'], b = date_2022), axis=1)
df['ISR'] = (df['Sueldo mensual'] * 1).apply(lambda x: ISR.function_isr(x))
df['IMSS_Ob'] = df['SBC'].apply(lambda x: CalculoIMSSObrero.function_imss_obrero(x))
df['IMSS_Pt'] = df.apply(lambda x: CalculoIMSSPatron.function_imss_patronal(x = x['SBC'], date=date_2022 ), axis=1)
df['ISN'] = df['Sueldo mensual'] * 0.03
_isr_ = df['Sueldo mensual'].apply(lambda x: ISR.function_isr(x))
df['Neto'] = df['Sueldo mensual'] - _isr_ - df['IMSS_Ob']
df['Aguinaldo'] = df.apply(lambda x: SBC.funcion_aguinaldo(sueldo = x['Sueldo mensual'], a = x['Fecha Ingreso'], b = date_2022), axis=1)
df['Prima Vacacional'] = df.apply(lambda x: SBC.funcion_pv(sueldo = x['Sueldo mensual'], date_ingreso = x['Fecha Ingreso'], date_base = date_2022), axis=1)
```

```python
incremento = 0.078

def fondo(x):
    datos = SBC()
    found = x * incremento/2
    limit = (datos._uma * 365 * 1.3) / 12
    if found < limit:
        return found
    else:
        return limit
        
def excedente_fondo(x):
    datos = SBC()
    found = x * incremento/2
    limit = (datos._uma * 365 * 1.3) / 12
    if found > limit:
        return (found - limit) * 2
    else:
        return 0
```

```python
df_fondo['fondo_t'] = df_fondo['Sueldo mensual'].apply(lambda x: fondo(x))
df_fondo['fondo_p'] = df_fondo['Sueldo mensual'].apply(lambda x: fondo(x))
df_fondo['excedente_fondo'] = df_fondo['Sueldo mensual'].apply(lambda x: excedente_fondo(x))
```

```python
def vales(sueldo):
    if (sueldo * 0.05) > 1500:
        return int(1500)
    else:
        return int(round(sueldo * 0.05, 0))
```

```python
df_fondo
```

```python
_base = df_fondo['Sueldo mensual'] + df_fondo['fondo_t'] + df_fondo['excedente_fondo']
```

```python
df_fondo['Vales'] = _base.apply(lambda x: vales(x))
```

```python
df_fondo['SD'] = _base.apply(lambda x: x / 30)
df_fondo['SBC'] = df_fondo.apply(lambda x: SBC.funcion_sbc(sueldo = x['Sueldo mensual'] + x['fondo_t'] + x['excedente_fondo'], vales = x['Vales'], a = x['Fecha Ingreso'], b = date_2022), axis=1)
df_fondo['ISR'] = _base.apply(lambda x: ISR.function_isr(x))
df_fondo['IMSS_Ob'] = df_fondo['SBC'].apply(lambda x: CalculoIMSSObrero.function_imss_obrero(x))
df_fondo['IMSS_Pt'] = df_fondo.apply(lambda x: CalculoIMSSPatron.function_imss_patronal(x = x['SBC'], date=date_2022 ), axis=1)
df_fondo['ISN'] = _base.apply(lambda x: x * 0.03)
_isr_ = _base.apply(lambda x: ISR.function_isr(x))
df_fondo['Neto'] = df_fondo['Sueldo mensual'] - _isr_ - df_fondo['IMSS_Ob']
df_fondo['Aguinaldo'] = df_fondo.apply(lambda x: SBC.funcion_aguinaldo(sueldo = x['Sueldo mensual'] + x['fondo_t'] + x['excedente_fondo'], a = x['Fecha Ingreso'], b = date_2022), axis=1)
df_fondo['Prima Vacacional'] = df_fondo.apply(lambda x: SBC.funcion_pv(sueldo = x['Sueldo mensual'] + x['fondo_t'] + x['excedente_fondo'], date_ingreso = x['Fecha Ingreso'], date_base = date_2022), axis=1)
df_fondo['Dias Vacaciones'] = df_fondo.apply(lambda x: SBC.funcion_dias_vacaciones( date_ingreso = x['Fecha Ingreso'], date_base = date_2022), axis=1)

df_fondo['Sueldo'] = df_fondo['Sueldo mensual'] + df_fondo['fondo_t'] + df_fondo['excedente_fondo']

isr_quincena = df_fondo['Sueldo mensual'].apply(lambda x: ISR.function_isr(x))

df_fondo['ISR quincena 1'] = isr_quincena.apply(lambda x: x/2)
df_fondo['ISR quincena 2'] = df_fondo['ISR'] - df_fondo['ISR quincena 1']


# df_fondo['Porcentaje'] = ((df_fondo['Sueldo'] + df_fondo['fondo_p']) - df_fondo['Sueldo mensual'])/df_fondo['Sueldo mensual']
# df_fondo['Porcentaje2'] = ((df_fondo['Sueldo']) - df_fondo['Sueldo mensual'])/df_fondo['Sueldo mensual']
```
